```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

def capture():
    path = "D:\\Projects\\ERB\\LH\\fgvc-aircraft-2013b.tar\\fgvc-aircraft-2013b\\data\\images\\"
    
    images=[]
    titles=[]
    i=0
    
    for filename in os.listdir(path):
        img = cv2.imread(os.path.join(path,filename), 1)
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img)
            titles.append(str(filename))
            i = i+1
        if i == 10:
            break
        
    
    for j in range(i):
        plt.subplot(1, i , j+1)
        plt.imshow(images[j])
        plt.title(titles[j])
        plt.xticks([])
        plt.yticks([])
    
    plt.show()
        
```

```python
capture()
```
