```python
path = "D:\\Projects\\Data\\unsupervised-image-clustering-master\\animal_images\\C1-Abyssinian\\"
img = cv2.imread(path + "Abyssinian_9.jpg", 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
plt.show()
```

```python
block_size = 255
constant = 2
```

```python
th1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)
```

```python
output = [img, th1, th2]
titles=['Original', 'Thresh Mean', 'Gaussian']
```

```python
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(output[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
